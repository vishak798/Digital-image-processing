```python
import cv2
from matplotlib import pyplot as plt
```

```python
image=cv2.imread('color_img.jpg',0)
```

```python
cv2.imshow('image',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.title('Histogram')
plt.hist(image.ravel(),256,[0,256])
plt.show()
```

```python
equalize=cv2.equalizeHist(image)
```

```python
import numpy as np
result=np.hstack((image,equalize))
```

```python
cv2.imshow('Equalized',equalize)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.title('Equalized histogram')
plt.hist(equalize.ravel(),256,[0,256])
plt.show()
```
