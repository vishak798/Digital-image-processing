```python
# RGB channel
import cv2
from google.colab.patches import cv2_imshow 
img=cv2.imread("image.jpg",1)
img = cv2.resize(img, (250, 230))
B, G, R = cv2.split(img)
cv2_imshow(img) 
cv2_imshow(R)  
cv2_imshow(G) 
cv2_imshow(B) 
#grayscale without built in function
grayimg = 0.5*R + 0.5*G + 0.5*B;
cv2_imshow(grayimg) 
```

```python
# image to binary image
import cv2 
import numpy as np 
img = cv2.imread('color_img.jpg',0) 
ret,img1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
img1 = cv2.resize(img1, (250, 230))
cv2.imshow("_",img1) 
```

```python
# histogram og grayscale
from matplotlib import pyplot as plt
img = cv2.imread('image.jpg', 0)
img = cv2.resize(img, (250, 230))
cv2_imshow(img) 
dst = cv2.calcHist(img, [0], None, [256], [0,256])
equ = cv2.equalizeHist(img)
cv2_imshow(equ) 
plt.hist(img.ravel(),256,[0,256])
plt.title('Histogram for gray scale image')
plt.show()
plt.hist(equ.ravel(),256,[0,256])
plt.title('Histogram for equalized image')
plt.show()
```
